# Random Forest Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Intuition

A random forest is an ensemble model, which is when you take multiple algorithms and combine them together to end up with better performing predictions than just a single model. This concept leverages the fact that correct predictions will reinforce each other, whilst errors will tend to cancel each other out. However, for this to work, the individual decision trees must not make the same errors; i.e. they must be uncorrelated. On the other hand, decision trees are prone to over-fitting.

To address these two problems, random forests use bootstrap aggregation (bagging) - where each individual decision tree is trained only on a subset of the data (and sometimes able to use only a subset of available predictors). This ensures that each decision tree will be different from the next and will prevent overfitting, whilst still retaining predictive power.

Ensemble models are also known to be more stable - a change in your dataset may change a single decision tree drastically, but will average out across `Ntree` trees.

How does it work?
1. Pick `K` random data points from your training set
2. Build a decision tree associated with these `K` data points (a subset of your training set)
3. Choose number `Ntree` of trees you want to build and repeat steps 1-2 `Ntree` times
4. For a new data point, make all `Ntree` trees predict the value of Y for the data point in question. Your prediction will be the average of all `Ntree` Y predictions

As an extension of this, when picking your `K` data points, also pick a subset `P` of the available predictor features and only train the individual tree using features available in `P`.


[Image source](https://www.tibco.com/reference-center/what-is-a-random-forest)

## Limitations

* Lack of interpretability: random forests are considered black boxes as it is difficult to visualise the aggregate effect of the sometimes thousand of individual trees they are made from. While the effect of each individual predictor may be visualised by the use of techniques such as [SHAP lines](https://shap.readthedocs.io/en/latest/example_notebooks/overviews/An%20introduction%20to%20explainable%20AI%20with%20Shapley%20values.html) and feature importance measures, it is hard to tell exactly how the random forest used each specific input to arrive at the answer. Hence, random forests are more useful in situations where accuracy of results is more important than interpretability.

## Dataset and business problem description

This dataset is the same as that used in the previous lesson; it contains job position levels for a certain company as well as the salaries for each level.

```python
dataset = pd.read_csv('./data/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values # only grab the Level column for X
y = dataset.iloc[:, -1].values

X, y
```

## Training the Random Forest Regression model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

## Predict salary

```python
regressor.predict([[6.5]])
```

## Visualising results

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y)
plt.plot(
    X_grid, 
    regressor.predict(X_grid),
    color = 'coral'
)
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')

# More trees = more splits, and therefore more steps
```
