# Decision Tree Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Intuition

CART: Classification and Regression Trees; here we'll be focusing on the regression trees. These are slightly more complex than classification trees.

We'll work with two independent variables, $x_1$ and $x_2$. We'll be working to predict $y$, which is on the third dimension which we won't be visualising yet.

Running your decision tree will split the scatterplot of $x_1$ and $x_2$ into segments in a process called recursive partitioning. How and where the splits are conducted are determined by the algorithm and involves looking at information entropy - it checks if adding each split will tell us something more about our data (this equates to reducing standard deviation in the predictions). Splits will continue to be added until a specified stopping criterion is reached, which could be:

* Stop when data points at each leaf are all of the same predicted category or value
* Stop when the leaf contains less than five data points
* Stop when further branching does not improve homogeneity beyond a minimum threshold

As recursive partitioning uses only the best variables to split on for the next split, non-significant variables will be filtered out. Further, since splits tend to occur around central values, decision trees are robust to outliers.

How do these splits help to determine the value of $y$ for given $x_1$ and $x_2$? Take the average of $y$ for all points within each terminal leaf and that will be the value assigned to any new point that falls into the same leaf. ## Limitations

* Instability: slight changes in data may trigger changes to each split in the tree, resulting in a completely different tree; in the same vein, decision trees are also vulnerable to overfitting
* Inaccuracy: using binary questions to split the data may not always lead to the most accurate predictions

To minimise these issues, one might implement bootstrap aggregating such as in random forests.

## Dataset and business problem description

This dataset is the same as that used in the previous lesson; it contains job position levels for a certain company as well as the salaries for each level.

```python
dataset = pd.read_csv('./data/Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values # only grab the Level column for X
y = dataset.iloc[:, -1].values

X, y
```

## Training the Decision Tree Regression model on the whole dataset

We will not need to implement feature scaling since decision trees are not sensitive to the range of data passed through it.

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state = 0) # random_state = 0 just fixes the random seed

regressor.fit(X, y)
```

## Predict salary

```python
regressor.predict([[6.5]])
```

## Visualising results

Note this algorithm isn't best adapted to datasets with only one independent variable. Further, since each terminal node uses the average result of $y$, we're left with a stepped function as seen below.

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y)
plt.plot(
    X_grid, 
    regressor.predict(X_grid),
    color = 'coral'
)
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
```
