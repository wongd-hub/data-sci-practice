## ML vs. DL vs. AI - What's the difference?

[Source](https://dropbox.com/s/biqmldsnmtp8g8s/OU2-Difference-Between-ML-DL-AI.pdf?dl=0)

### Artificial Intelligence

* Science empowering computers to mimic human intelligence such as in decision making, text processing, and visual perception. This is a broad field, of which Machine Learning and Deep Learning are subsets.

### Machine Learning

* Subfield of AI that enables machines to imrpove at a given task with experience.
* Subfields of ML include supervised learning (training algorithms on labelled data - classification and clustering), unsupervised learning (training on unlabelled data - clustering), and reinforcement learning (algorithms that take actions to maximise a cumulative reward).

### Deep Learning

* A subfield of ML that relies on the training of Deep Artificial Neural Networks (ANNs), which process information with a model inspired by how the human brain works.
* The human brain consists of neurons that communicate with each other, whereas an ANN consists of artificial neurons in a multi-layered mathematical model.
* The differentiation here is an ANN's ability to extract features automatically.
    1. Machine learning process:
        1. Select model to train
        2. Manually perform feature extraction
        
    2. Deep learning process:
        1. Select architecture of network
        2. Features are automatically extracted by feeding in the training data along with the target class
        
## Regression types

### Simple Linear Regression

* Predicting a dependent variable, y, using a single independent variable, x.
* Assumes a linear relationship between x and y.
* Aims to optimise the following relationship: $y = b + mx$
    * Sidenote, [good MathJax (Jupyter Notebook math renderer) here](https://math.meta.stackexchange.com/questions/5020/mathjax-basic-tutorial-and-quick-reference).

### Multiple Linear Regression

* Extends simple linear regression to > 1 independent variables: $y = b_0 + m_1x_1 + m_2x_2 + \dots + m_nx_n$

### Polynomial Regression

* Models the relationship between y and independent variable x as an $n^{th}$ degree polynomial.

    $y = b_0 + b_1x + b_2x^2 + \dots + b_nx^n$
