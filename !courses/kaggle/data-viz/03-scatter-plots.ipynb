Scatter plots in `seaborn` use the `scatterplot` function. 

## Qualifying Performance

Qualifying is a 3-tiered event that occurs on the Saturday of each race weekend, where drivers compete to set the fastest laps on track. The fastest driver will start from first position on race day (Sunday); so on and so forth. 

All 20 drivers compete in the first segment of qualifying (called Q1), and the 5 slowest drivers are eliminated. Q2 then commences and the next 5 slowest drivers are eliminated. Q3 is the final segment, where the starting grid is set for the last 10 drivers.

We'd generally expect drivers to get better over the period of qualifying as they get more comfortable with the track and the track surface gets 'rubbered in' allowing for more tyre grip. However, this is not always the case, as sudden inclement weather or car issues (damage or mechanical failures) can cause slower lap times in later qualifying segments.

The best lap times set by drivers in Q3 vs their best time in Q1 are plotted below, accompanied by a 45-degree line. Points below the line represent drivers improving on their Q1 time in Q3; points above the line indicate the opposite. Times are coloured based on the track they were set on, as lap times will naturally vary based on the length and features of the track.

Note that we could draw a regression line through the points with `sns.regplot` (with `sns.lmplot` being an option to plot regression lines by data subset), but the 45 degree line illustrates what we're looking for better in this case.

```python
qualy_subset = qualifying[qualifying.gp_name.isin(rd.sample(qualifying['gp_name'].unique().tolist(), 5))]

# Plot
plt.figure(figsize = (14, 6))
plt.title(f"Driver Lap Times in Q1 vs. their Q3 Times")

ax = sns.scatterplot(data = qualy_subset, x = 'q1', y = 'q3', hue = 'gp_name')
ax.axline(xy1 = (70, 70), slope = 1, linewidth = 1, color = 'gray')

plt.legend(title = 'Grand Prix', loc = 'upper left')
plt.xlabel('Q1 Best Time (Seconds)')
plt.ylabel('Q3 Best Time (Seconds)')
```

## Swarm Plots

We can visualise how best Q3 times per driver differ across different  in another way, using `seaborn`'s `swarmplot` (a 'categorical' scatterplot, ie. a box-plot alternative).

```python
warnings.filterwarnings('ignore')

plt.figure(figsize = (14, 6))
plt.title(f"Driver's best Q3 times by track")
sns.swarmplot(data = qualy_subset, x = 'gp_name', y = 'q3')

plt.xlabel('Grand Prix')
plt.ylabel('Best Q3 Time (Seconds)')
```
